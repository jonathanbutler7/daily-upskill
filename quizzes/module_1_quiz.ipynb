## Q1: Data Transformations (Generators & Comprehensions)

Given a list of log entries (strings), create a generator that filters for "ERROR" logs and yields a dictionary with the keys `timestamp` and `message`.

**Example input:** `"2026-03-30 10:00:00 ERROR: Connection failed"`  
**Example output:** `{'timestamp': '2026-03-30 10:00:00', 'message': 'Connection failed'}`

```python
def parse_errors(log_lines):
    # Your code here
    for line in log_lines:
        if "ERROR" in line:
            # Hint: split the line by " ERROR: "
            raise NotImplementedError()

# Test data
logs = [
    "2026-03-30 10:00:00 INFO: System start",
    "2026-03-30 10:01:00 ERROR: Database unreachable",
    "2026-03-30 10:02:00 DEBUG: Heartbeat",
    "2026-03-30 10:05:00 ERROR: Out of memory"
]
```

```python
results = list(parse_errors(logs))
assert len(results) == 2
assert results[0] == {'timestamp': '2026-03-30 10:01:00', 'message': 'Database unreachable'}
assert results[1]['message'] == 'Out of memory'
print("Q1 Success!")
```

## Q5: Decorators

Implement a decorator `ensure_int` that checks if the first argument to a function is an integer. If not, raise a `TypeError`.

```python
import functools

def ensure_int(func):
    @functools.wraps(func)
    def wrapper(*args, **kwargs):
        # Your code here
        raise NotImplementedError()
    return wrapper

@ensure_int
def double(x):
    return x * 2
```

## Q6: Protocols (Structural Typing)

Define a `Drawable` protocol that requires a `draw` method which takes no arguments and returns a `str`.

```python
from typing import Protocol, runtime_checkable

@runtime_checkable
class Drawable(Protocol):
    # Your code here
    raise NotImplementedError()

class Circle:
    def draw(self) -> str:
        return "Drawing a circle"

def render(item: Drawable):
    return item.draw()
```
